Mutacja: dodanie losowego prostokąta do zbioru, a usunięcie takich z którymi on koliduje

Krzyżowanie: krzyżowanie względem losowej cięciwy koła - powyżej cięciwy zachowujemy prostokąty ojca, a poniżej te z prostokątów matki, które z nimi nie kolidują

Każde rozwiązanie po mutacji lub krzyżowaniu jest poprawiane poprzez sciągnięcie wszystkich prostokątów maksymalnie w lewo, następnie rozrost (dodawanie nowych prostokątów stykających wierzchołkiem się z istniejącym) w prawo, a następnie ściągnięcie w dół i rozrost w górę.

Inicjalizacja osobnika jest wypełnieniem koła losowymi prostokątami i poprawieniem tego osobnika.

Selekcja opiera się na elitaryzmie i selekcji ruletkowej.    

Udało mi się zaimplementować ciekawy działający algorytm. Niestety nie zdążyłem wykonać testów dla wszystkich zbiorów. Poniżej załączam ilustracje działania algorytmu dla jednego wybranego zbioru. W raporcie znajdzie się dokładny opis algorytmu oraz testy działania dla wszystkich zbiorów. 

# Testowanie dla poszczególnych zbiorów

# r = 800

```python
n_iters = 250
n_epochs = 4
population_size = 100
n_mutations = 32
n_crossings = 70
elite_count = 10
e = Evolution(population_size, 800, "../data/cutting/r800.csv")
e.best().plot()
print(e.best().evaluate())
loss_history_x = np.array([])
loss_history_y = np.array([])
for epoch in range(n_epochs):
    for i in range(n_iters):
        e.iter(n_mutations, n_crossings, elite_count)
        loss_history_x = np.concatenate((loss_history_x, np.repeat(epoch * n_epochs + i, population_size)))
        loss_history_y = np.concatenate((loss_history_y, np.array(list(map(lambda x: x.evaluate(), e.population)))))
    print("epoch " + str(epoch) + ": ")
    plt.scatter(loss_history_x, loss_history_y, alpha=0.3)
    e.best().plot()
    print(e.best().evaluate())
```
