# kwadratowa

```python
a = np.array([1, 1, 2])
f = lambda x: np.sum((a * x) ** 2)
min_bound = [-1] * 3
max_bound = [1] * 3
size = 200
```

## testowanie różnych wariancji mutacji

```python
population = Population(f, size, min_bound, max_bound)
loss_history_x = np.array([])
loss_history_y = np.array([])
sigma = 0.1
for i in range(2000):
    population.iteration(60, sigma, 140, 32, 4)
    loss_history_x = np.concatenate((loss_history_x, np.repeat(i, size)))
    loss_history_y = np.concatenate((loss_history_y, np.array(list(map(lambda x: x.eval(), population.individuals)))))

plt.scatter(loss_history_x, np.log(loss_history_y), alpha=0.3)
plt.scatter(loss_history_x, np.log(loss_history_y), alpha=0.3)
plt.xlabel("Iteracja")
plt.ylabel("logarytm funkcji przystosowania")
plt.suptitle("Wartości przystosowania osobników populacji w czasie")
plt.title("sigma = " + str(sigma))
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')
x = np.arange(-2, 2, 0.01)
y = np.arange(-2, 2, 0.01)
X, Y = np.meshgrid(x, y)
Z = 20 + X ** 2 + Y ** 2 - 10 * np.cos(2 * np.pi * X) - 10 * np.cos(2 * np.pi * Y)
surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
fig.colorbar(surf, shrink=0.5, aspect=8)
plt.show()
```

```python
A = 10
f = lambda x: 10 * len(x) + np.sum(x ** 2 - A * np.cos(2 * np.pi * x))
min_bound = [-20] * 5
max_bound = [20] * 5
size = 200
```

```python
population = Population(f, size, min_bound, max_bound)
loss_history_x = np.array([])
loss_history_y = np.array([])
best_history = population.select_best().values
sigma = 0.1
for i in range(6000):
    best_history = np.vstack([best_history, population.select_best().values])
    population.iteration(60, sigma, 140, 32, 4)
    loss_history_x = np.concatenate((loss_history_x, np.repeat(i, size)))
    loss_history_y = np.concatenate((loss_history_y, np.array(list(map(lambda x: x.eval(), population.individuals)))))

plt.scatter(loss_history_x, np.log(loss_history_y), alpha=0.3)
plt.scatter(loss_history_x, np.log(loss_history_y), alpha=0.3)
plt.xlabel("Iteracja")
plt.ylabel("logarytm funkcji przystosowania")
plt.suptitle("Wartości przystosowania osobników populacji w czasie")
plt.title("sigma = " + str(sigma))
plt.show()
for i in range(5):
    plt.plot(np.log(np.abs(best_history[:, i])), label = "log (" + str(i+1) + " współrzędnej)")
plt.legend()
plt.xlabel("Iteracja")
plt.title("sigma = " + str(sigma))
plt.show()
```
